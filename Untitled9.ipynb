```python
class Car:
    def __init__(self, model, name, year):
        self.model = model
        self.name = name 
        self.year = year
    def show_car(self):
        return f"{self.model} {self.name} {self.year}"
```

```python
my_car = Car("tata" , "safari" , 2022)
```

```python
print(my_car.model)
```

```python
print(my_car.show_car())
```

```python
sai_car = Car("toyota","innova",2013)
sai_car.model
```

```python
print(sai_car.show_car())
```

```python
class Bat(Car):
    def __init__(self,model,name,year,battery):
        super().__init__(model,name,year)
        self.battery = battery
```

```python
sai_car = Bat("toyota","innova",2013,"80mvh")

print(sai_car.battery)
```

```python
print(sai_car.show_car())
```
