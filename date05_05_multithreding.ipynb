```python
import time
import threading #python module 
start = time.perf_counter()

def test_func():
    print("do something")
    print("sleep for 1 sec")
    time.sleep(2)
    print("done with sleeping")

# run the program on two thread
t1 = threading.Thread(target = test_func) # to start the thread
t2 = threading.Thread(target = test_func)

t1.start()
t2.start()
end = time.perf_counter()

print(f"the program finshed in { round(end-start ,2)} seconds ")


# t1 and t2 running paralaerly but other code than this threads running paralally too
# to solve this problem we have to use join
```

```python
# Join use >> join execute the mentioned thread and then the main thread will be executed
def test_func():
    print("do something")
    print("sleep for 1 sec")
    time.sleep(2)
    print("done with sleeping")

# run the program on two thread
t1 = threading.Thread(target = test_func) # to start the thread
t2 = threading.Thread(target = test_func)

t1.start()
t2.start()

t1.join() 
t2.join()
end = time.perf_counter()

print(f"the program finshed in { round(end-start ,2)} seconds ")

# now this thread will be completed in 2 sec's
```

```python
import time
import threading

start = time.perf_counter()
def count_time():
    print("this is first function")
    print("this is second function")
    time.sleep(2)
    print("this is third fucntion ")
    print("all function completed")

thread= []
for i in range(10):
    t = threading.Thread(target = count_time)
    t.start()
    t.join()
    thread.append(t)

end = time.perf_counter()

print(f" this is total time : { round(end-start ,2) } seconds")

# This function took 20 sec because t.join is runnig inside the loop 
```

```python
import time
import threading

start = time.perf_counter()
def count_time():
    print("this is first function")
    print("this is second function")
    time.sleep(2)
    print("this is third fucntion ")
    print("all function completed")

thread= []
for i in range(10):
    t = threading.Thread(target = count_time)
    t.start()
    thread.append(t)

for thread in thread:
    thread.join()

end = time.perf_counter()

print(f" this is total time : { round(end-start ,2) } seconds")

#using the theread join to run every therad parally this code will complete excution in 2 sec
```

```python
# if you want to pass argument in fuunc
import time
import threading

start = time.perf_counter()
def count_time(args):
    print("this is first function")
    print(f"this is second {args} function")
    time.sleep(args)
    print("this is third fucntion ")
    print("all function completed")

thread= []
for i in range(10):
    t = threading.Thread(target = count_time , args =[2])
    t.start()
    thread.append(t)

for thread in thread:
    thread.join()

end = time.perf_counter()

print(f" this is total time : { round(end-start ,2) } seconds")

#using the theread join to run every therad parally this code will complete excution in 2 sec
```

```python
# use cases 
# multi therading works well with I/O bound task >> means where some output has to wait for input
# Example >> reading-writting , network communincation , sql quiries
```

```python
# https://github.com/itsfoss/text-script-files

import time
import threading

start = time.perf_counter()
url_list = [
    'https://github.com/itsfoss/text-script-files/blob/cb07dab8c39d2c97308ec95b80815b84b8d77f89/agatha.txt'
    'https://github.com/itsfoss/text-script-files/blob/cb07dab8c39d2c97308ec95b80815b84b8d77f89/agatha_complete.txt'
    'https://github.com/itsfoss/text-script-files/blob/cb07dab8c39d2c97308ec95b80815b84b8d77f89/sample_log_file.txt'
    'https://github.com/itsfoss/text-script-files/blob/cb07dab8c39d2c97308ec95b80815b84b8d77f89/sherlock.txt'
]

data_list = ['data1.txt','data2.txt','data3.txt','data4.txt']

import urllib.request

def file_download(url, filename):
    urllib.request.urlretrieve(url, filename)

thread = []
for i in range(len(url_list)):
    t = threading.Thread(target = file_download , args = (url_list[i], data_list[i]))
    t.start()
    thread.append(t)
for thread in thread:
    thread.join()

end = time.perf_counter()

print(f" time to finish the code is {round(end-start,2)} seconds")
```
